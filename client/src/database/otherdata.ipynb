```python
## generate userA

import pandas as pd
from faker import Faker
import random
import string

fake = Faker()

def generate_password(lenn):
    return  ''.join(random.choice(string.ascii_letters + string.digits) for _ in range(lenn))

# Generate data for 100 users
num_users = 100
usernames = [fake.user_name() for _ in range(num_users)]
names = [fake.name() for _ in range(num_users)]
passwords = [generate_password(random.randint(8, 20)) for _ in range(num_users)]
regions = [fake.country() for _ in range(num_users)]
ages = [random.randint(18, 65) for _ in range(num_users)]
languages = [fake.language_name() for _ in range(num_users)]
platforms = [random.choice(['WIN', 'MAC', 'LNX']) for _ in range((num_users))]

# Create a dataframe with the generated data
df = pd.DataFrame({
    'username': usernames,
    'name': names,
    'password': passwords,
    'region': regions,
    'age': ages,
    'language': languages,
    'platform': platforms
})

df.to_csv('userA.csv')

with open("userA.sql", 'w') as file:
    file.write("INSERT INTO userA (username, name, password, region, age, language, platform) VALUES\n")
    for i, row in df.iterrows():
        file.write(f"('{row['username']}', '{row['name']}', '{row['password']}', '{row['region']}', {row['age']}, '{row['language']}', '{row['platform']}'),\n")
```

```python
# generate post
from faker import Faker
import random

fake = Faker()

num_of_posts = 500

# Generate 100 fake post titles and texts
titles = [fake.sentence(nb_words=random.randint(3, 10)) for _ in range(num_of_posts)]
post_texts = [fake.paragraph(nb_sentences=random.randint(1, 5)) for _ in range(num_of_posts)]

# Generate the post data
post_data = []
for i in range(num_of_posts):
    username = random.choice(usernames)
    gid = random.randint(1, 153)
    title = titles[i]
    post_text = post_texts[i]
    post_data.append((username, gid, title, post_text))

# Create the post dataframe
post_df = pd.DataFrame(post_data, columns=['username', 'gid', 'title', 'postText'])

post_df.to_csv('post.csv')

with open("post.sql", 'w') as file:
    file.write("INSERT INTO post (username, gid, title, postText) VALUES\n")
    for i, row in post_df.iterrows():
        file.write(f"('{row['username']}', {row['gid']}, '{row['title']}', '{row['postText']}'),\n")
```

```python
## generate comment

import random
from faker import Faker

fake = Faker()

# Generate comment data
num_comments = 2000
comments = []
for i in range(num_comments):
    comment_body = fake.paragraph(nb_sentences=random.randint(1, 5))
    post_id = random.randint(1, num_of_posts)
    username = random.choice(usernames)
    comments.append((comment_body, post_id, username))
    
# Create DataFrame
comment_df = pd.DataFrame(comments, columns=['comment_body', 'post_id', 'username'])

comment_df.to_csv('comment.csv')

with open("comment.sql", 'w') as file:
    file.write("INSERT INTO comment (comment_body, post_id, username) VALUES\n")
    for i, row in comment_df.iterrows():
        file.write(f"('{row['comment_body']}', {row['post_id']}, '{row['username']}'),\n")
```

```python
# Generate like data
num_likes = 2000
likes = set()
while len(likes) < num_likes:
    post_id = random.randint(1, num_of_posts)
    user_id = random.randint(1, num_users)
    likes.add((post_id, user_id))
    
# Convert set to list and create DataFrame
likes_df = pd.DataFrame(list(likes), columns=['post_id', 'user_id'])

comment_df.to_csv('likes.csv')

with open("likes.sql", 'w') as file:
    file.write("INSERT INTO likes (post_id, user_id) VALUES\n")
    for i, row in likes_df.iterrows():
        file.write(f"({row['post_id']}, {row['user_id']}),\n")
```
